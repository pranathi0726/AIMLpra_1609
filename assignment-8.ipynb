```python
car_train = train_data.drop('binaryClass', axis=1)
car_labels = train_data[['binaryClass']]
```

```python
from sklearn.preprocessing import OrdinalEncoder

from sklearn.pipeline import Pipeline

pipe = Pipeline([('ord_enc', OrdinalEncoder())])

car_train_prepared = pipe.fit_transform(car_train)
```

```python
from sklearn.preprocessing import LabelEncoder

label_enc = LabelEncoder()

car_labels_prepared = label_enc.fit_transform(car_labels)
```

```python
car_train_prepared.shape, car_labels_prepared.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier()
tree_clf.set_params(class_weight='balanced', criterion='entropy')

tree_clf.fit(car_train_prepared, car_labels_prepared)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20,15))
plot_tree(tree_clf, feature_names=car_train.columns,
          class_names=['N','P']);
```

```python
from sklearn.metrics import accuracy_score, classification_report,\
confusion_matrix

preds = tree_clf.predict(car_train_prepared)
print(f"Accuracy:{accuracy_score(car_labels_prepared, preds)}")
print(f"ClassReport:{classification_report(car_labels_prepared,preds)}")
print(f"Confusion Matrix:{confusion_matrix(car_labels_prepared,preds)}")
```

```python
from sklearn.metrics import confusion_matrix
```

```python
def accuracy(features, model, labels):
    predictions = model.predict(features)
    acc = np.mean(predictions == labels)
    print(f"Accuracy: {acc}")
```

```python
def conf_matrix(features, model, labels):
    """
    Generates and prints the confusion matrix.

    Args:
    features (array-like): The features of the dataset.
    model: The trained machine learning model.
    labels (array-like): The true labels of the dataset.
    """

    predictions = model.predict(features)
    cm = confusion_matrix(labels, predictions)  # Assuming confusion_matrix is from sklearn
    print("Confusion Matrix:")
    print(cm)
```
